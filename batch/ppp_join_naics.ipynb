```python
import pandas as pd
```

```python
READ_PATH_PPP = "/Users/jon/Desktop/data-cleaning-project/data/PPP Data up to 150k 080820 HI.csv"
READ_PATH_NAICS = "/Users/jon/Desktop/data-cleaning-project/data/6-digit_2017_Codes.csv"
ppp_df = pd.read_csv(READ_PATH_PPP)
naics_df = pd.read_csv(READ_PATH_NAICS)
```

```python
ppp_df = ppp_df.astype({"NAICSCode": "Int32"})
ppp_df = ppp_df.astype({"NAICSCode": "string"})
ppp_df.head()
```

```python
#naics_df = naics_df.dropna()
naics_df = naics_df.astype({"2017 NAICS Code":"int32","2017 NAICS Title":"string"}, errors='ignore')
naics_df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1, inplace=True)
naics_df.head()
```

```python
joined_ppp = ppp_df.merge(
    right=naics_df,
    how='left',
    left_on='NAICSCode',
    right_on='NAICS17'
)
```
